```python
#importing Libraries
import pandas as pd
import re
import cleantext
import empath
from cleantext import clean
lexicon = empath.Empath()
```

```python
#Empty List for storing values for each Teams
Pos=[]
Neg=[]
Neu=[]
```

```python
#clean the text 
#Created a function to clean the tweets
def cleanTxt(text):
    text = cleantext.clean(text, no_emoji=True)
    text = re.sub(r'@[A-Za-z0-9]+','', text) #Removed @mentions
    text = re.sub(r'#','',text) # Removing the '#' symbol
    text = re.sub(r'RT[\s]+', '',text) #Removing RT
    text = re.sub(r'https?:\/\/\S+', '',text) #Remove the hyper link
    return text
```

```python
#How Empath library shows results
data=pd.read_csv('hallabol.csv')
analysis=lexicon.analyze(data['Text'][1], normalize=True)
analysis
```

```python
#Sentiment Analysis on Rajasthan Royals 
data=pd.read_csv('hallabol.csv')
positive = 0
negative = 0
neutral = 0

#Appending to same column
data['Text'] = data['Text'].apply(cleanTxt)

for i in data['Text']:
    
    analysis=lexicon.analyze(i,categories=["cheerfulness", "pride",  "superhero", "surprise", "leader", "heroic", "celebration", "love", 
                                  "strength", "power", "achievement", "positive_emotion","winner","aggression"
                                          ,"weakness", "fear", "anger", "pain", "negative_emotion","loser"], normalize=True)
    if analysis["cheerfulness"] !=0.0 or analysis["pride"] !=0.0 or analysis["superhero"]!=0.0 or analysis["surprise"]!=0.0 or analysis["leader"] !=0.0 or analysis["heroic"] !=0.0 or analysis["love"] !=0.0 or analysis["power"] !=0.0 or analysis["positive_emotion"] !=0.0 or analysis["celebration"] !=0.0 or analysis["achievement"]!=0.0 or analysis["strength"]!=0.0 or analysis["winner"]!=0.0:
        positive += 1
        
    elif analysis["aggression"] !=0.0 or analysis["weakness"] !=0.0 or analysis["fear"] !=0.0 or analysis["anger"] !=0.0 or analysis["pain"] !=0.0 or analysis["negative_emotion"] !=0.0 or analysis["loser"] !=0.0:
        negative += 1
    else:
        neutral += 1
        
print("Rajasthan Royals Positive Reviews:",positive)
print("Rajasthan Royals Negative Reviews:",negative)
print("Rajasthan Royals Neutral Reviews:",neutral)     

Pos.append(positive)
Neg.append(negative)
Neu.append(neutral)
```
